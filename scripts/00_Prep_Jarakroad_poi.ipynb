```python
# Impor GeoPandas
import geopandas as gpd

# Membaca file GeoJSON
print("1. Membaca data...")
buildings = gpd.read_file('Building_Footprint_Cleaned_fix.geojson')
roads = gpd.read_file('Main_Road_Cleaned_fix.geojson')
pois = gpd.read_file('POI_Cleaned_fix.geojson')

# Menyamakan Sistem Koordinat (CRS) ke dalam METER
print("2. Menyamakan koordinat ke satuan meter (EPSG:3857)...")
buildings_m = buildings.to_crs(epsg=3857)
roads_m = roads.to_crs(epsg=3857)
pois_m = pois.to_crs(epsg=3857)

# Menghitung Jarak Terdekat
print("3. Menghitung jarak ke jalan utama dan POI terdekat (mohon tunggu sebentar)...")

# Fungsi ini mencari jarak dari SATU bangunan ke SEMUA jalan, lalu mengambil nilai minimumnya (terdekat)
buildings_m['jarak_jalan_m'] = buildings_m.geometry.apply(lambda x: roads_m.distance(x).min())

# Lakukan hal yang sama untuk POI (Pusat Ekonomi)
buildings_m['jarak_poi_m'] = buildings_m.geometry.apply(lambda x: pois_m.distance(x).min())

# Mengembalikan koordinat ke bentuk asal
buildings_final = buildings_m.to_crs(epsg=4326)

# Simpan hasil akhir
print("4. Menyimpan data hasil ukuran...")
buildings_final.to_file('Building_Footprint_Distance.geojson', driver='GeoJSON')

print("Selesai! File 'Building_Footprint_Distance.geojson' sudah jadi.")
```
